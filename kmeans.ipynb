```
import numpy as np
import cv2
```

```
img=cv2.imread('/content/gdrive/My Drive/data/data1contrast.jpg')
```

```
# contrast
img_lab=cv2.cvtColor(img,cv2.COLOR_BGR2LAB)
l,a,b=cv2.split(img_lab)
# equ=cv2.equalizeHist(l)
# ui=cv2.merge((equ,a,b))
# hi=cv2.cvtColor(ui,cv2.COLOR_LAB2BGR)
clahe=cv2.createCLAHE(clipLimit=3.0,tileGridSize=(8,8))
cl=clahe.apply(l)
ui=cv2.merge((cl,a,b))
hi=cv2.cvtColor(ui,cv2.COLOR_LAB2BGR)
cv2.imwrite('/content/gdrive/My Drive/data/sev3_c3_contrast.jpg',hi)
```
